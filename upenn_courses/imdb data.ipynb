```python
import pandas as pd
import numpy as np
xls = pd.ExcelFile('imdb.xlsx')
df = xls.parse('imdb')
df_directors = xls.parse('directors')
df_countries = xls.parse('countries')

df = pd.merge(left=df, right=df_countries, 
              how='inner', left_on='country_id', 
              right_on='id')

df = pd.merge(left=df, right=df_directors, 
              how='inner', left_on='director_id', 
              right_on='id')



df['movie_title'] = df['movie_title'].str.replace('Ê', "")
df.head()
```

```python
# clean director data
df_directors_clean = df_directors.drop_duplicates("id")

df_directors_clean
```

```python
# most movies director
director_with_most = df["director_name"].value_counts().loc[lambda x: x == x.max()]
director_with_most
```

```python
# all movies of the director

atts = ["movie_title", "imdb_score"]
all_movies_ratings = df[df['director_name'] == str(director_with_most.index[0])][atts]
all_movies_ratings.head()
```

```python
# random choose a movie

good = df['imdb_score'] > 8.3

goodmovie = df[good]

import random
rand = random.randint(0, len(goodmovie) - 1)
rand_goodmovie = goodmovie[rand:rand + 1]
random_title = rand_goodmovie['movie_title']
print(random_title)
random_imdb_score = rand_goodmovie['imdb_score']
print(random_imdb_score)
```

```python
# summary statistics for imdb_score and gross
score_gross_description = df[['imdb_score','gross']].describe()

nolan = df['director_name'].isin(['Christopher Nolan'])

nolan_mean = df[nolan]['imdb_score'].mean()
```
