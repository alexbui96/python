```python
import csv
import os
filepath = os.getcwd() + '/data/' + "ufo-sightings.csv"

ufosightings = [] 

with open(filepath, 'r') as csvfile:
    reader = csv.DictReader(csvfile)

    for row in reader:
        ufosightings.append(row)

ufosightings_count = len(ufosightings)

# Sightings were there in the US
sightings_us = [row for row in ufosightings if row["country"] == "us"]

# Find the sighting(s) that lasted more than ten seconds.
def is_valid_duration(duration_as_string):
    try:
        duration = float(duration_as_string)
    except ValueError:
        return False
    else:
        return duration > 10

fball = [row for row in sightings_us if row["shape"] == "fireball" and is_valid_duration(row["duration (seconds)"])]

# the datetime and duration of the longest sighting
fball_max = max(fball, key = lambda x: float(x["duration (seconds)"]))

# sorted fball by duration (seconds) in reverse order
fballsorted = sorted(fball, key = lambda x: float(x["duration (seconds)"]), reverse = True)

# the state had a longest lasting "fireball"
state = fball_max["state"]

# Find the sighting(s) that lasted more than 0 second.
def is_significant_sighting(duration_as_string):
    try:
        duration = float(duration_as_string)
    except ValueError:
        return False
    else:
        return duration > 0

valid_ufosightings = [row for row in ufosightings if is_significant_sighting(row["duration (seconds)"])]

# the shortest sighting
min_sighting = min(valid_ufosightings, key = lambda x: float(x["duration (seconds)"]))
min_duration = float(min_sighting["duration (seconds)"])

sightings_shapes = [row["shape"] for row in ufosightings]

count = {}

for i in range(len(sightings_shapes)):

    shape = sightings_shapes[i]

    if not(shape in count):

        count[shape] = count.get(shape, 1) 
    
    else:
        count[shape] += 1

sorted_count = sorted(count.items(), key= lambda x: x[1], reverse = True)

top3shapes = sorted_count[0:3]
```
